```python
import os
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler

np.random.seed(42)

PROCESSED_DIR = r"C:\Project_EquiGuard\data\processed"
MODELS_DIR    = r"C:\Project_EquiGuard\outputs\models"
ARRAYS_DIR    = os.path.join(PROCESSED_DIR, "arrays")

print("Libraries loaded.")
```

```python
# Load engineered feature matrices

X_train = np.load(os.path.join(ARRAYS_DIR, "X_train_final.npy"))
y_train = np.load(os.path.join(ARRAYS_DIR, "y_train.npy"))
X_val   = np.load(os.path.join(ARRAYS_DIR, "X_val_final.npy"))
y_val   = np.load(os.path.join(ARRAYS_DIR, "y_val.npy"))
X_test  = np.load(os.path.join(ARRAYS_DIR, "X_test_final.npy"))
y_test  = np.load(os.path.join(ARRAYS_DIR, "y_test.npy"))

print(f"X_train: {X_train.shape}")
print(f"X_val:   {X_val.shape}")
print(f"X_test:  {X_test.shape}")

print(f"\ny_train: {np.unique(y_train, return_counts=True)}")
print(f"y_val:   {np.unique(y_val,   return_counts=True)}")
print(f"y_test:  {np.unique(y_test,  return_counts=True)}")
```

```python
# Verify missing values

print(f"Nulls in X_train: {np.isnan(X_train).sum()}")
print(f"Nulls in X_val:   {np.isnan(X_val).sum()}")
print(f"Nulls in X_test:  {np.isnan(X_test).sum()}")
```

```python
# Fit scaler on training data and transform all splits

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled   = scaler.transform(X_val)
X_test_scaled  = scaler.transform(X_test)

print("Scaling complete.")
print(f"\nX_train mean after scaling: {X_train_scaled.mean():.4f}")
print(f"X_train std after scaling:  {X_train_scaled.std():.4f}")

joblib.dump(scaler, os.path.join(MODELS_DIR, "scaler.pkl"))
print(f"\nScaler saved to: {MODELS_DIR}/scaler.pkl")
```

```python
# Save scaled datasets
np.save(os.path.join(ARRAYS_DIR, "X_train_scaled.npy"), X_train_scaled)
np.save(os.path.join(ARRAYS_DIR, "X_val_scaled.npy"),   X_val_scaled)
np.save(os.path.join(ARRAYS_DIR, "X_test_scaled.npy"),  X_test_scaled)

print("Scaled arrays saved.")
print(f"\nFinal shapes:")
print(f"X_train_scaled: {X_train_scaled.shape}")
print(f"X_val_scaled:   {X_val_scaled.shape}")
print(f"X_test_scaled:  {X_test_scaled.shape}")
print(f"\ny_train: {y_train.shape}")
print(f"y_val:   {y_val.shape}")
print(f"y_test:  {y_test.shape}")
print("\nPreprocessing completed.")
```
